## 2.4.5: Random Walks

```python
#Start with a small sample/program to test that your program works

import matplotlib.pyplot as plt
import numpy as np
#Start at origin, take 100 steps, and each step is sampled on the standard normal distribution
#Assume that x-displacement and y-displacement are independent

#Generating the displacements in a 2x5 table of elements
delta_X=np.random.normal(0,1,(2,5))
#take the first component of each vector as the x-component, and the second as the y
#graphing the displacements that we generated from the normal distribution
plt.plot(delta_X[0], delta_X[1], 'go')
#now we need a cumulative sum of the elements as they go along the row:
X = np.cumsum(delta_X, axis=1)
# Take a look at the X vs delta_X:                                                                                   
# In [8]: X                                                                                                                                  
# Out[8]: 
# array([[ 0.15744803, -0.61294002, -1.10072224, -0.43525636,  1.20699334],
#        [-1.12769682, -1.51363123, -1.16334405, -0.64103523, -0.24131312]])

# In [9]: delta_X                                                                                                                            
# Out[9]: 
# array([[ 0.15744803, -0.77038805, -0.48778222,  0.66546588,  1.6422497 ],
#        [-1.12769682, -0.38593442,  0.35028719,  0.52230882,  0.39972211]])

#put different pieces together


```

```python
#The actual program
import numpy as np
import matplotlib.pyplot as plt

X_0 = np.array([[0],[0]])
delta_X = np.random.normal(0,1,(2,100))
# X = np.cumsum(delta_X, axis = 1) this was put into the X = np.concatenate function
X = np.concatenate((X_0, np.cumsum(delta_X, axis = 1)), axis = 1)
plt.plot(X[0], X[1], 'ro-')
plt.savefig('rw3.pdf')
```

```python
#3d random walker
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

ax = plt.gca(projection="3d")

X_0 = np.array([[0],[0],[0]])
delta_X = np.random.normal(0,1,(3,100))
# X = np.cumsum(delta_X, axis = 1) this was put into the X = np.concatenate function
X = np.concatenate((X_0, np.cumsum(delta_X, axis = 1)), axis = 1)
ax.scatter(X[0],X[1],X[2], c='r',s=100)
ax.plot(X[0],X[1],X[2], color='r')
```
