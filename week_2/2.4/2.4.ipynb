### Example 1: Roll a die 100 Times and plot a histogram of the outcomes (How frequent the numbers from 1 to 6 appear in 100 samples)
***

```python
#Ex: Roll a die 100 Times and plot a histogram of the outcomes (How frequent the numbers from 1 to 6 appear in 100 samples)
import numpy as np
import matplotlib.pyplot as plt
import random
random.choice(range(1,7))
rolls = []
for k in range(100):
    rolls.append(random.choice(range(1,7)))
plt.hist(rolls, bins= np.linspace(0.5,6.5,7)); #start is 0.5, end is 6.5 and since i want 6 bins, I need 7 points
```

### Why should we expect this result?
- Theorem of Large Numbers (Probability)
    - We should expect about the same number of 1s, 2s, 3s,... $\because$ they all have the same probability
    - Let's rerun the example with 10000 repetitions


```python
for k in range(10000):
    rolls.append(random.choice(range(1,7)))
plt.hist(rolls, bins= np.linspace(0.5,6.5,7)); #start is 0.5, end is 6.5 and since i want 6 bins, I need 7 points
```

This is a lot more even than the first example. If we raise the number of rolls even higher ($10^6$), we'll get an even smoother plot:


```python
for k in range(1000000):
    rolls.append(random.choice(range(1,7)))
plt.hist(rolls, bins= np.linspace(0.5,6.5,7)); #start is 0.5, end is 6.5 and since i want 6 bins, I need 7 points
```

### Example 2: Rolling 10 independent dice ($x_1$ to $x_{10}$)
***
- New variable y (which has the sum of all of the $x$ variables: $y=\sum_{i=1}^{10} x_i$)
- We want to understand the distribution of the variable y by simulating its values many times, then plotting a histogram
- Before starting, let's try to anticipate what the histogram might look like
    - Useful to do in general
    - We're adding 10 $x$s:
        - the min value of each $x$ is 1 ($y$ value is 10)
        - the max value of each $x$ is 6 ($y$ value is 60)
    - Think about value of max value: for y to be 60, each x needs to be at its max value (unlikely). It is much more likely that it will take some intermediate value (e.x. 30)
    - We expect the histogram to peak at the center, and thin out towards the ends
    

```python
ys =[]
for rep in range(100):
    y=0
    for k in range(10):
        x=random.choice(range(1,7))
        y+=x
    ys.append(y)
```

```python
#Check whether the bounds are within expected bounds
print(min(ys))
print(max(ys))
```

```python
plt.hist(ys);
```
